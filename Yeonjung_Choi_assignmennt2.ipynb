# Data Analysis

4.Analyse the energy intensity, defined as energy consumption per unit of GSP, for each
state annually and chart these figures

This problem is to analyze the annual energy intensity of each state and draw related graphs. According to the assignment requirements, energy intensity is equal to energy consumption divided by DSP.there is no colum called"energy intensity",it need to be calculated by the formula below:

energy intensity(GJ/milliom)=energy consumption(PJ)/GSP(million)

input=energy intensity(PJ/million)
output=broken line graph

```python
# Convert energy consumption columns to numeric
energy_columns = ['Coal(PJ)', 'Oil(PJ)', 'Gas(PJ)', 'Renewables(PJ)']
df[energy_columns] = df[energy_columns].apply(pd.to_numeric, errors='coerce')

# Calculate total energy consumption for each state annually
df['Total_Energy_Consumption'] = df[energy_columns].sum(axis=1)

# Calculate Gross State Product (GSP) for each state annually
# For simplicity, let's assume GSP remains constant for each state across the years
gsp_per_state = df.groupby('State')['GSP($ milion)'].mean()

# Calculate energy intensity (PJ/$ million) for each state annually
df['Energy_Intensity'] = df['Total_Energy_Consumption'] / df['State'].map(gsp_per_state)

# Plotting
plt.figure(figsize=(10, 6))
for state, data in df.groupby('State'):
    plt.plot(data['Year'], data['Energy_Intensity'], label=state)

plt.title('Energy Intensity (PJ/$ million) by State Annually')
plt.xlabel('Year')
plt.ylabel('Energy Intensity (PJ/$ million)')
plt.xticks(rotation=45)  # Corrected the typo in rotation parameter
plt.legend()
plt.grid(True)
plt.show()
```
